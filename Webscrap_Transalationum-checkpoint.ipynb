```python
base_url = 'http://www.unesco.org/xtrans/bsresult.aspx?a=&stxt=&sl=egy&l=eng&c=&pla=&pub=&tr=&e=&udc=&d=&from=&to=&tie=a'

urls =[]
for frame in range(10,68,10):
    urls.append(base_url + f'&fr={frame}')
```

```python
urls = urls + [base_url]
```

```python
session = FuturesSession(max_workers=10)
futures = [session.get(url) for url in urls]
list_of_pages = [future.result() for future in futures]
```

```python
frames = []

for page in list_of_pages:
    soup = BeautifulSoup(page.text, "lxml")
    results_table = soup.find('table', {'class':'restable'})
    books = results_table.find_all('td', {'class':'res2'})
    json = []
    for book in books:
        json.append({span['class'][0] : span.text 
                 for span in book.find_all('span')})
    df = pd.DataFrame.from_records(json)  
    frames.append(df)
```

```python
df = pd.concat(frames, ignore_index=True, sort=False)
```
