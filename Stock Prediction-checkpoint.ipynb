```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
```

```python
# Data
index = pd.read_csv('c:/Sekolah/nestle.csv')
print(index.tail())
```

```python
# Adjusting close price
index = index[['Adj Close']]
# New data
print(index.tail())
```

```python
# A variable for predicting 'n' days out into the future
forecast = 30 
#Create another column (the target or dependent variable) shifted 'n' units up
index['Prediction'] = index[['Adj Close']].shift(-forecast)
#print the new data set
print(index.tail())
```

```python
# Make independent data set(X) and convert to numpy arry
X= np.array(index.drop(['Prediction'],1))

# Remove the last 'n' rows
X=X[:-forecast]
print(X)
```

```python
# Make dependent data set (y) and convert to numpy array
y= np.array(index['Prediction'])
y=y[:-forecast]
print(y)
```

```python
# split the data
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Make and train SVM (Regressor)
svr_rbf = SVR (kernel='rbf', C=1e3, gamma=0.1)
svr_rbf.fit(x_train,y_train)
```

```python
# Testing Model SVM
svm_confidence = svr_rbf.score(x_test, y_test)
print('svm confidence: ', svm_confidence)
```

```python
# Make and train Linear Regression
lr= LinearRegression()

# Train the model
lr.fit(x_train, y_train)
```

```python
# Testing Model
lr_confidence = lr.score(x_test, y_test)
print('lr confidence :', lr_confidence)
```

```python
# Set x_forecast equal to the last 30 rows of the original data set from Adj. Close column
x_forecast = np.array(index.drop(['Prediction'],1))[-forecast:]
print(x_forecast)
```

```python
# Prediction for Linear Regression
lr_prediction = lr.predict(x_forecast)
print(lr_prediction)

# Prediction for SVM
svm_prediction = svr_rbf.predict(x_forecast)
print(svm_prediction)
```
