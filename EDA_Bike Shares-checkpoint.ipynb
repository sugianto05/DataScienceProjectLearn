```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bike = pd.read_csv('c:/Sekolah/london_merged.csv')
bike.head()
```

```python
bike['timestamp'] = pd.to_datetime(bike['timestamp'])
bike['year'] = bike['timestamp'].dt.year
bike['month'] = bike['timestamp'].dt.month
bike['day'] = bike['timestamp'].dt.day
bike['weekday'] = bike['timestamp'].dt.weekday
```

```python
# 0.0 = non holiday
# 1.0 = holiday
bike['is_holiday'].value_counts()
```

```python
# 0.0 = Not weekend
# 1.0 = Weekend
bike['is_weekend'].value_counts()
```

```python
# 0.0 = spring
# 1.0 = summer
# 2.0 = fall
# 3.0 = winter
bike['season'].value_counts()
```

```python
#1.0 = Clear
#2.0 = few cloud
#3.0 = broken clouds
#4.0 = cloudy
#7.0 = Rain
#10  = rain with thunderstorm
#26  = snowfall
bike['weather_code'].value_counts()
```

```python
hard_bike_season=bike[(bike['season']==3) & (bike["weather_code"]==26)]
hard_bike_season
```

```python
# the most hard time for bike shares due to winter and snowfall
hard_bike_season['timestamp'].count()
```

```python
#find null
bike.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
bike['season']= bike['season'].map({0:'spring', 1:'summer', 2:'autumn', 3:'winter'})
bike['weather'] = bike['weather_code'].map({1:'Clear', 2:'Few Clouds', 3:'Broken Clouds', 4:'Cloudy', 7:'Rain', 10:'Stormy', 26:'Snow', 94:'Freezing Fog'})
bike['is_holiday'] = bike['is_holiday'].map({0:'Non Holiday', 1:'Holiday'})
bike['is_weekend'] = bike['is_weekend'].map({0:'Non Weekend', 1:'Weekend'})
bike = bike.drop(['weather_code'],axis=1)

#write a summary of the data 
bike.describe()
```

```python
f, ax = plt.subplots(2,2,figsize=(15, 12))

sns.barplot(x='weather',y='cnt', data=bike, order=['Clear','Few Clouds', 'Broken Clouds', 'Cloudy','Rain', 'Stormy', 'Snow','Freezing Fog'], ax=ax[0][0])
sns.barplot(x='season',y='cnt', data=bike, ax=ax[0][1])
sns.barplot(x='is_weekend',y='cnt', data=bike, ax=ax[1][0]) 
sns.barplot(x='is_holiday',y='cnt', data=bike, ax=ax[1][1]) 

for i in range(len(ax)):
    for k in range(2):
        plt.setp(ax[i][k].xaxis.get_majorticklabels(), rotation=45, ha="right", size = 12)
        plt.setp(ax[i][k].yaxis.get_majorticklabels(), size = 12)
        ax[i][k].xaxis.label.set_fontsize(20)
        ax[i][k].yaxis.label.set_fontsize(20)

        plt.tight_layout()
```

```python
f, ax = plt.subplots(2,2,figsize=(15, 12))
sns.lineplot(data=bike,x='hum', y='cnt',ax=ax[0][0])
sns.lineplot(data=bike,x='wind_speed', y='cnt',ax=ax[0][1])
sns.lineplot(data=bike,x='t1', y='cnt',ax=ax[1][0])
sns.lineplot(data=bike,x='t2', y='cnt',ax=ax[1][1])

for i in range(len(ax)):
    for k in range(2):
        plt.setp(ax[i][k].xaxis.get_majorticklabels(), size = 12)
        plt.setp(ax[i][k].yaxis.get_majorticklabels(), size = 12)
        ax[i][k].xaxis.label.set_fontsize(20)
        ax[i][k].yaxis.label.set_fontsize(20)
plt.tight_layout()
```

```python
bins = np.linspace(min(bike["hum"]), max(bike["hum"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
#elaborate more humid column by divide into 3 categories
bike['hum-binned'] = pd.cut(bike['hum'], bins, labels=group_names, include_lowest=True )
bike[['hum','hum-binned']].head(10)
```

```python
bike["hum-binned"].value_counts()
```

```python
plt.bar(group_names, bike["hum-binned"].value_counts())

# set x/y labels and plot title 
plt.xlabel('Humid')
plt.ylabel("Count")
plt.title("Humidity Categories")
```
